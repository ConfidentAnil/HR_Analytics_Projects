```python
import pandas as pd
import numpy as np
import math
import scipy.stats as st
import statistics as s

#Finding the upper and lower interval.
#Formula : u/l interval = xbar +- (z-value * std.error)
#xbar = sample_mean
#z-value = standard z value for normally distributed data and not zscore of the sample data.
#std.error = standard error, which is given as sigma/squareroot_of_sample_size
##**********For calculating upper and lower interval, the sample data should be normally distributed data, because the zvalue relis on the the normally distributed data only*************
#Creating random values which are normally distributed

#creating a normally distributed data i.e., 
data = np.random.normal(size = 100)
#This is the observered value, which is the actual value we have




#value : xbar, sample mean
xbar = data.mean()
print(f"Sample Mean or Xbar: {xbar}")


#Z-value has the standard values for normally distributed data, standard values meaning, fix values for each confidence level there is a fixed z-value:
#We are going to find the standard z-value at 95% confidence level.
z_value = st.norm.ppf(1-0.05)
print(f'Standarized Z_value at 95% confidence level : {z_value}')


#value : std.error, we use sem method from scipy.stats library.class
std_error = st.sem(data)
print(f'Standard_error of sample : {std_error}')


#Finding upper and lower interval

#upperinterval = xbar + (z_value * std_error) #Upper interval formula.
upper_interval = xbar + (z_value*std_error)


#lower_interval = xbar - (zvalue * std_error) #lower interval formula.
lower_interval = xbar - (z_value*std_error)


print(f'Lower Interval : {lower_interval}, Upper Interval : {upper_interval}')
#We can take the inference from this lower and upper intervals that the mean for the other samples will lie between (-0.2107562294693877, 0.10602492532707425) 95% times, or 95 times out of 100 times.


#To test the inference that population mean would fall between Lower Interval : -0.1522525323861959, Upper Interval : 0.18991204895690666
#creating a another sample size 100000 and checking if the mean value of this dataset will find in between the upper and lower interval.
data1 = np.random.normal(size = 100000)
print(f'Sample 2nd | (sample_mean) xbar : {data1.mean()}')
```

```python
#T-Test : 1 sample t-test, 2 sample t-test, paired t-test

print("T-test has types types : \n1)1 sample t-test\n2)2 sample t-test\n3)paired t-test")

#-------------------------------------------------1 Sample T-Test--------------------------------------------------

#Find T-test for question : The government says average height of boys in class 10th is 165cm

#Formulating hypothesis: H0 - The average height of boys in class 10th is 165cm
                        #H1 - The average height of boys in class 10th is not 165cm

from scipy.stats import ttest_1samp

from random import sample

#generating sample from values between 145cm to 180, sample size(n) = 30
sample1 = sample(range(130,170), 30)


#For finding 1sample t-test, we use ttest_1samp method from scipy.stats library.class.
result_1samplettest = ttest_1samp(sample1, popmean=165)


print(f'The 1 sample t-test values are : {result_1samplettest}')

#As my pvalue is greater than 0.05, i.e. p-value > 0.05, we do not have enough evidence to reject null hypothesis H0, hence
#average height of boys in class 10th is 165cm



#---------------------------------------2 Sample T-Test----------------------------------------


#Find the 2 sample t-test for question : Is the average speed of car and bike same, here we have two population, p1 = car, p2 = bike

#Formulating hypothesis : H0 - The average speed of bike and car is same
                         #H1 - The average speed of car and bike is not same.
    
sample_b = sample(range(40,80), 30)
sample_c = sample(range(47,90),30)

#importing ttest_ind method from scipy.stats library.stats to find the 2 sample t-test
from scipy.stats import ttest_ind


print(f"The 2 sample t-test values are : {ttest_ind(sample_b, sample_c)}")

#The pvalue is 0.0056980115142678065 which is greater than 0.05 it means we do not have enough evidence to reject the H0
#Hence, our inference is average speed of car and bike is same.
#**Our DF here will be 58 and not 59 or 29, because we add the two samples df, 1st sample df is 29 + 29 of sample 2 = 58 Need to google this



#---------------------------------------------Paired T-Test------------------------------------------------------


#Find the Paired Sample T-test for question : The average income of employes from technical department in Google is same

#Formulating the hypothesis : H0 - The means of two samples will be same
                             #H1 - The means of two samples will not be same

    

sample_a = sample(range(300000, 450000), 30)
sample_b = sample(range(300000, 450000), 30)


#To find the 2 sample t-test we will use ttest_rel method from scipy.stats library.class
from scipy.stats import ttest_rel


#result
print(f"The paired sample t-test values are : {ttest_rel(sample_a, sample_b)}")

#The pvalue is less than 0.05, i.e. pvalue < 0.05 we have the enough evidence to reject the null hypothesis.
#Hence, the reference is the average income of employess from technical department in Google for two samples will not be same.
```

```python
#F-test
#We calculate F-statistics which compares the two variances of 2 samples
#And P value tells us if the two compares follow the same distribution or not, if yes, we accept the nully hypothesis otherwise reject it

#Formula in math
#f-test = sigma 1/sigma 2



#Formulating the hypothesis
#H0 : The variances are same
#H1 : The variances are not same

import scipy.stats as st

samplef1 = sample(range(0,100),30)
samplef2 = sample(range(0,100),30)

#Checking the variances are same, comparing them
f = np.var(samplef1)/np.var(samplef2)

#Finding p value using scipy.stats.f.cdf library.class.method.method. i.e., scipy.stats.f.cdf() to check if 2 samples are following same distribution.
p_value = st.f.cdf(f, len(samplef1)-1, len(samplef2)-1)
print(f'The F-statistics is :{f}\nP_value is :{p_value}')

#As the pvalue is greater than 0.05, we do not have enough evidence to reject the H0, i.e., null hypothesis.
#Hence the variance are same.

```

```python
#ANOVA Test : 1 way ANOVA, 2 way ANOVA

#In 1 way ANOVA test, we compare both means and variance, and there could be two or more samples and we find the f-statistics to find the 
#And just in F-test and T-test, the sample size <= 30

#Formulating hypothesis: H0 - The impact of enhancers are same
                        #H1 - The impact of enhancers are not the same.

#Generating four enhancers speed records.     
sampleanova1 = sample(range(130,250), 30)
sampleanova2 = sample(range(120,230), 30)
sampleanova3 = sample(range(100,260), 30)
sampleanova4 = sample(range(50,200), 30)

#Using f_oneway method from scipy.stats library.class to find the pvalue
from scipy.stats import f_oneway
f_oneway(sampleanova1, sampleanova2, sampleanova3, sampleanova4)
print('The pvalue is : 2.509162581992803e-10\nThe statistics value is : 19.550877126144158')
#As the pvalue is less than 0.05 we do not have enough evidence to accept the null hypothesis.



#------------------------------------------------------2 way ANOVA------------------------------------------------------
#In 2 way ANOVA, it is checked if the 2 factors have relationship with response variable, i.e., marks and 2 factors are Tuition and Lecture,
#There are Daily and Weekly classes.
#Formulating the hypothesis : H0 - There is no significance of two factors with response variable
                             #H1 - There is a significance of two factors with response variable

#importing s
import statsmodels.api as sm
from statsmodels.formula.api import ols
sample2wayanova1 = {"Lectures" : np.repeat(['Daily', 'Weekly'], 20),
                    "Tuition" : np.repeat(['Daily', 'Weekly'], 20),
                    "Marks" : sample(range(35,100),40)
                   }
dataframe = pd.DataFrame(sample2wayanova1)
dataframe

#By using the OLS, we are finding the relationship between marks with tuition and lectures
relationship = ols("Marks ~ C(Lectures)+C(Tuition)+C(Lectures):C(Tuition)", data = dataframe).fit()
relationship

#Using the OLS derived value which is finding the linear relationship between 2 facts and 1 response variable with error component
#Supplying the OLS derived value into sm.stats.anova_lm()
sm.stats.anova_lm(relationship)


#The conclusion is as pvalue is <= 0.05, we do not have enough evidence to reject the H0
#Hence there is no effect or impact of tuition and lectures with marks variable. Even if the tuition and lectures are not taken, the marks will remain the same. 
#It means there is no significance and marks is a independent variable.
```

```python
#Ztest : one sample Z-test
#We can perform Ztest when sample size is greater than 30 and in our one sample z-test
#Find the value to check if average weight of a teenager pre covid was 55kg what is the weight now?

#Formulating the hypothesis : H0 - The average weight of a teenager is 55kg
                             #H1 - The average weight of a teenager is not 55kg

#generating sample size > 30 of weights from population.
samplez = sample(range(26,86),60)

#importing statsmodels.stats.weightstats.ztest, i.e. library.class.method.method 
from statsmodels.stats.weightstats import ztest
ztest(samplez, value = 55)

print(f'The pvalue of 1 sample z-test is : 0.045945088346883475\nThe stats of 1 sample z-test value is : 1.9958974315773466')

#Since Pvalue is 0.045945088346883475 which is less than 0.05, we reject the null hypothesis.
#Hence The average weight of a teenager is not 55kg


#--------------------------------------------------2 sample z-test---------------------------------------------------------
#Picking another sample of size 60 from the same population

#Formulating the hypothesis : H0 - The sample1 and sample2 weights are same
                             #H1 - The sample1 and sample2 weights are not the same.

samplez2 = sample(range(29,89),60)

#Using same ztest method from statsmodels.stats.weightstats library.class.method to find the 2 sample z-test, here we will add two samples from the population.
ztest(samplez, samplez2)

print(f'The pvalue of 2 sample z-test is : 0.34676888345244306\nThe stats of 2 sample z-test value is : -0.94087507228077')

#Since pvalue is greater than 0.05, we do not have enough evidence to reject the null hypothesis
#Hence sample 1 and sample 2 weights are same.
```

```python
#Chi Square Test

#In Chi Square test we find the relationship between categorical variable. If the variable is a continuous variable we can convert it into categorical variable
#Reading dataframe from purchase.csv and performing Chi-Square test and other analysis

#Is purchase category indepdent of product_category_1 category.

#Formulating the hypothesis : H0 - Purchase and product_category_1 are not related, they are independent
                             #H1 - Purchase and product_category_1 are related, they are dependent
purchasedf = pd.read_csv('purchase.csv')
purchasedf.head()

#Converting continuous variable into categorical variable. 
#1. Finding the maximum value of Purhcase column.
purchasedf['Purchase'].max()
#2. And using pd.cut() method to cut or group the 0 to 10000, as 0 and 10000 to maximum value i.e. 23961 as 1
purchasedf['Purchase'] = pd.cut(purchasedf['Purchase'], bins = [0,10000,23961], labels = [0,1])

#We are going to get a cross table/frequency table for purchase and product_category_1 columns.
cross_table = pd.crosstab(purchasedf['Purchase'], purchasedf['Product_Category_1'])
cross_table

#validating the frequency of product category 0 has value 26131 for product_category_1 = 1, When purchase is 0 and product_category_1 the value is 26131, validate this.
len(purchasedf[(purchasedf['Purchase'] == 0) & (purchasedf['Product_Category_1'] == 1)])

#Running a chi-square statistics find if there is a relationship between product_category_1 and purchase. Are they independent of each other or not.
#cross_table dataframe becomes the input to the chi-square statistics of Observed value. This observed value go and sit in the xsquare = summation(observedi -expectedi)square/expectedi. Expected i is calculated under the assumption of independence.

#importing chi2_contingency from scipy.stats i.e., library.class to find the chi-square statistics and pvalue
from scipy.stats import chi2_contingency
result = chi2_contingency(cross_table)


#The expected values for product_category_1 = 1 to 20 when purchase = 0. These are the expected values which indicate complete indepences, since they are not the same values as our observed, hence they are dependent variables.
result[3][0]

#The expected values for product_category_1 1 t0 20 when purchase = 1. These are the expected values which indicate complete indepences, since they are not the same values as our observed, hence they are dependent variables.
result[3][1]

#Since our pvalue is less than 0.05, we do not have enough evidence to accept the null hypothesis.
#Hence our two variables are related to each other. That means the product_category_1 has relationship with purchase, for maybe product_category_1 = 12, the purhcase value is x, it is x and not less or greater than x hence they are related.
```
